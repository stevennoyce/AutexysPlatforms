```python
from scipy.optimize import curve_fit

def trapFill1(t, v, A, tau):
    return v + A*np.exp(-t/tau)

popts = []

for parameters in parametersSets:
    timestamps = np.array(parameters['Results']['timestamps'])
    timestamps = timestamps - timestamps[0]
    timestamps = timestamps/3600
    Id = np.array(parameters['Results']['id_data'])
    
    popt, pcov = curve_fit(trapFill1, timestamps, abs(Id))
    
    popts.append(popt)
    
    timestamps2 = np.linspace(0,2,100)
    plt.plot(timestamps2, trapFill1(timestamps2, *popt))
    plt.plot(timestamps, abs(Id))

plt.show()

taus = [popt[2] for popt in popts]
plt.semilogy(taus)
plt.ylabel('Tau [hours]')
plt.show()

print(np.median(taus))
```

```python
IdsInitial = []
IdsFinal = []
Vgss = []

for parameters in parametersSets:
    Id = np.array(parameters['Results']['id_data'])
    Vgs = np.array(parameters['Results']['vgs_data'])
    IdsInitial.append(np.mean(Id[0:2]))
    IdsFinal.append(np.mean(Id[-60:-1]))
    Vgss.append(Vgs[0])

fig = plt.figure(figsize=(1.7,1.89))
colors = dpu.colorsFromMap('plasma', 0, 0.75, 2)['colors']
plt.semilogy(Vgss, np.abs(IdsInitial), label='Initial\n   DR > {:.0f}\n'.format(round(IdsInitial[-1]/IdsInitial[0]/1000)*1000), c=colors[0])
plt.semilogy(Vgss, np.abs(IdsFinal), label="Terminal\n   DR $\\approx$ {:.0f}".format(IdsFinal[-1]/IdsFinal[0]), c=colors[1])
plt.legend()
plt.xlabel('$V_{GS}^{{Hold}}$')
plt.ylabel('$I_D$ [A]')
plt.tight_layout()
plt.savefig(saveFolder + '/Figure 5c - Initial and Terminal Responses.png', transparent=True, dpi=300)
plt.show()
```

```python
IdsInitial = []
IdsFinal = []
Vgss = []
IdsigmasInitial = []
IdsigmasFinal = []

for parameters in parametersSets:
    Id = np.array(parameters['Results']['id_data'])
    Vgs = np.array(parameters['Results']['vgs_data'])
    startIds = Id[0:30]
    IdsFinal.append(np.median(Id[-100:-1]))
    startIdsl = np.polyval(np.polyfit(range(len(startIds)), startIds, 2), range(len(startIds)))
    startIdsml = np.array(startIds) - startIdsl
#     plt.plot(startIds)
#     plt.plot(startIdsl)
#     plt.plot(startIdsml)
#     plt.show()
    IdsInitial.append(startIdsl[0])
    IdsigmasInitial.append(np.std(startIdsml))
    IdsigmasFinal.append(np.std(Id[-100:-1]))
    Vgss.append(Vgs[0])

IdsInitial = np.array(IdsInitial)
IdsFinal = np.array(IdsFinal)

stepHi =(IdsInitial[:-1] - IdsInitial[1:])
stepHf = (IdsFinal[:-1] - IdsFinal[1:])

plt.plot(IdsInitial)
plt.plot(IdsFinal)
plt.show()

plt.plot(stepHi)
plt.plot(stepHf)
plt.show()

SNRi = stepHi/IdsigmasInitial[1:]
SNRf = stepHf/IdsigmasFinal[1:]

# fig = plt.figure(figsize=(1.7,1.89))
colors = dpu.colorsFromMap('plasma', 0, 0.75, 2)['colors']
plt.plot(Vgss[1:], SNRi, label='Initial\n   DR > {:.0f}\n'.format(round(IdsInitial[-1]/IdsInitial[0]/1000)*1000), c=colors[0])
plt.plot(Vgss[1:], SNRf, label="Terminal\n   DR $\\approx$ {:.0f}".format(IdsFinal[-1]/IdsFinal[0]), c=colors[1])
plt.legend()
plt.xlabel('$V_{GS}^{{Hold}}$')
plt.ylabel('$I_D$ [A]')
plt.tight_layout()
# plt.savefig(saveFolder + '/Figure 5c - Initial and Terminal Responses.png', transparent=True, dpi=300)
plt.show()
```
