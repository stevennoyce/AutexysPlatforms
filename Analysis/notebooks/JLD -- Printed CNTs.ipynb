```python
data_30uL  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=4, maxExperiment=4)
data_40uL  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=6, maxExperiment=6)
data_50uL  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=8, maxExperiment=8)
data_75uL  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=10, maxExperiment=10)
data_100uL = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=11, maxExperiment=11)
data_125uL = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=12, maxExperiment=12)

data_25uL_PBS  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=20, maxExperiment=20)
data_50uL_PBS  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=21, maxExperiment=21)
data_75uL_PBS  = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=22, maxExperiment=22)
data_100uL_PBS = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=23, maxExperiment=23)
data_125uL_PBS = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=24, maxExperiment=24)
data_150uL_PBS = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=25, maxExperiment=25)
data_175uL_PBS = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=26, maxExperiment=26)
data_200uL_PBS = dlu.loadSpecificDeviceHistory('../../AutexysData/jay/Quartz/JP3/A/5-6', 'DrainSweep.json', minExperiment=27, maxExperiment=27)

data_all_H2O = [data_30uL,data_40uL,data_50uL,data_75uL,data_100uL,data_125uL]
data_firsts_H2O = [data_set[0] for data_set in data_all_H2O]

data_all_PBS = [data_25uL_PBS,data_50uL_PBS,data_75uL_PBS,data_100uL_PBS,data_125uL_PBS,data_150uL_PBS,data_175uL_PBS,data_200uL_PBS]
data_firsts_PBS = [data_set[0] for data_set in data_all_PBS]

# === Plot 2-terminal leakage curves ===

fig2, (ax2,) = dpu.makeDevicePlot('OutputCurve', data_firsts_H2O, data_firsts_H2O[0]['Identifiers'], mode_parameters={'enableLegend':False, 'enableColorBar':True, 'sweepDirection':'forward', 'publication_mode':False})
dpu.saveExternalPlot(fig2, 'DI_Water_Leakage_vs_Volume_Curves')

fig3, (ax3,) = dpu.makeDevicePlot('OutputCurve', data_firsts_PBS, data_firsts_PBS[0]['Identifiers'], mode_parameters={'enableLegend':False, 'enableColorBar':True, 'sweepDirection':'forward', 'publication_mode':False})
dpu.saveExternalPlot(fig3, 'PBS_Leakage_vs_Volume_Curves')

# === Plot drain-source leakage vs. DI water volume ===

drain_current_peaks = [0]
drain_current_means = [0]
drain_current_std = [0]
for data_segment in data_all_H2O:
    peaks = [max(deviceRun['Results']['id_data'][0]) for deviceRun in data_segment]
    drain_current_peaks.append(peaks)
    drain_current_means.append(np.mean(peaks))
    drain_current_std.append(np.std(peaks))

PBS_current_peaks = [0]
PBS_current_means = [0]
PBS_current_std = [0]
for data_segment in data_all_PBS:
    peaks = [max(deviceRun['Results']['id_data'][0]) for deviceRun in data_segment]
    PBS_current_peaks.append(peaks)
    PBS_current_means.append(np.mean(peaks))
    PBS_current_std.append(np.std(peaks))

fig1, (ax1) = dpu.makeBlankPlot(figsize=(2,2))
ax1.set_xlabel('DI $H_2O$ Volume (L)')
ax1.set_ylabel('2-terminal Leakage (nA)')
ax1.errorbar([0,30,40,50,75,100,125], np.array(drain_current_means) * 1e9, yerr=np.array(drain_current_std) * 1e9, color='#351996', linewidth=1, marker='o', ms=4, label='DI Water', capsize=2, capthick=0.5, elinewidth=0.5,)
ax1.errorbar([0,25,50,75,100,125,150,175,200], np.array(PBS_current_means) * 1e9, yerr=np.array(PBS_current_std) * 1e9, color='#851996', linewidth=1, marker='o', ms=4, label='PBS', capsize=2, capthick=0.5, elinewidth=0.5,)
ax1.legend(title="$V_{DS} = 1.0$ V ", loc="upper left")
#ax1.set_xticks([0,25,50,75,100,125])
dpu.saveExternalPlot(fig1, 'DI_Water_Leakage_vs_Volume')


```
